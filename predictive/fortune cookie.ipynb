### SPACY ANALYSE

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_md')
```

```python
fortunecookie = open("fortunecookie.txt").read()
```

```python
doc = nlp(fortunecookie)
```

```python
from collections import Counter
```

```python
word_counts = Counter([item.text for item in doc if item.pos_ == "ADJ"])
```

```python
word_counts.most_common(10)
```

```python
total_words = sum(word_counts.values())
[(item[0], word_counts[item[0]] / total_words) for item in word_counts.items()][:25]
```

```python
unique_words = list(set([item.text for item in doc if item.is_alpha]))
```

```python
from math import e
square_weirdness = [(item, pow(word_counts[item]/total_words, 2) / e**nlp.vocab[item].prob) for item in unique_words]
```

```python
square_weirdness
```

```python
[item[0] for item in sorted(square_weirdness, reverse=True, key=lambda x: x[1])][:15]
```

```python
noun_counts = Counter([item.text for item in doc if item.pos_ == 'NOUN'])

top_ten_nouns = [item[0] for item in noun_counts.most_common(20)]
top_ten_nouns
```

```python
def flatten_subtree(st):
    return ''.join([w.text_with_ws for w in list(st)]).strip()

prep_phrases = []
for word in doc:
    if word.dep_ == 'prep':
        prep_phrases.append(flatten_subtree(word.subtree))
        
precount = Counter([item for item in prep_phrases])
precount.most_common()
```

```python
subjects = []
for word in doc:
    if word.dep_ in ('nsubj', 'nsubjpass'):
        subjects.append(flatten_subtree(word.subtree))
        

subjectcount = Counter([item for item in subjects])
subjectcount.most_common()
```
